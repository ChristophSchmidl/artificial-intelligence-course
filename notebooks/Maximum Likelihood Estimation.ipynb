# Maximum Likelihood Estimation

Maximum Likelihood Estimation (MLE) is a common method for estimating parameters in a statistical/probabilistic model. In other words, you simply find the parameters that maximize the likelihood of observing what actually happened. In other works, which parameters explain my data the best?

Often with MLE when multiplying probabilities, you will take the logarithm of the likelihood and maximize that. This does not change the maximum and it prevents you from multiplying many small numbers, which is going to be problematic with finite precision. To prove that the maximum does not change, take a look at the following code snippet.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
%matplotlib inline

# Define your parameters for the normal distribution
mu = 0
variance = 1
sigma = math.sqrt(variance)

# Define the range and how many datapoints you want to have
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
# Plot the normal distribution with mu = 0 with maximum
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()

# Show that the log of the normal distribution
# has the same maximum
plt.plot(x, np.log(stats.norm.pdf(x, mu, sigma)))
plt.show()

# Sometimes people take the negative log of a function
# to keep the maximum/minimum but make the function
# convex for several optimization techniques.
# Then the maximization problem becomes a 
# minimization problem
plt.plot(x, -np.log(stats.norm.pdf(x, mu, sigma)))
plt.show()
```
